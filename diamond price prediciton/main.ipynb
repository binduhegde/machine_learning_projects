order: D = best and J = worst 
    
ascii value of J: 74. If we want J to be the worst, we have to assign it the number 1 and increase the number as the alphabet approaches D
So we're subtracting 73 from every letter. But for the letter I, the ascii value is also 73 so 73-73=0 but we want the number 2. That's why we put a condition there

```python

def assign_color(string):

    no = abs(ord(string) - 73)
    if no == 0: return 2
    return no

```

```python
df['color'] = df['color'].apply(assign_color)
```

```python
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import Sequential, callbacks
from tensorflow.keras.losses import MeanSquaredLogarithmicError
```

```python
early_stopping = callbacks.EarlyStopping(
                min_delta=0.001, 
                patience = 10,
                restore_best_weights=True
                )
```

Starting with the Dense units as 9, 18, 18, 1 because 9 is the number of features. 2 hidden layers with 9*2 units and the output layer has 1 unit which is the prediction value. So 9, 18, 18, 1 is a good place to start

```python
model = Sequential()

model.add(Dense(units=9, activation ='relu', input_dim=9))

model.add(Dense(units=18, activation ='relu'))

model.add(Dense(units=18, activation ='relu'))

model.add(Dense(units=1, activation ='linear'))
```

```python
# compiling the model
msle = MeanSquaredLogarithmicError()
model.compile(optimizer= 'adam',
             loss = msle,
             metrics = msle)
```

```python
# training the model

history = model.fit(xtrain, ytrain, 
         batch_size=10,
         validation_split=0.2,
         epochs = 100, 
        callbacks=[early_stopping])
```
