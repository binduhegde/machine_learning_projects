```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
models = {'Linear Regression': LinearRegression(), 
         'Decision Tree': DecisionTreeRegressor(),
         'Random Forest': RandomForestRegressor()}
```

```python
def train_models(x, y):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=2)

    result = {}
    
    for name, model in models.items():
        model.fit(xtrain, ytrain)
        pred = model.predict(xtest)
        acc = round(metrics.r2_score(pred, ytest), 2)
        result[name] = acc
    
    return result
        
```

```python
x = df[['Duration']]
y = df[['Calories']]

only_duration = train_models(x, y)
only_duration
```

```python
results_dict = {'Only Duration Column':only_duration,
               'Only relevant Columns': only_relevant,
               'All the Columns': all_columns}
results = pd.DataFrame(results_dict)
results
```

```python

# Reshape the DataFrame into long format
results_melted = results.reset_index().melt(id_vars='index', var_name='columns', value_name='values')

# Plotting the barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='index', y='values', hue='columns', data=results_melted, palette='Set2')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.title('Comparison of Different Results', fontsize=14)
plt.legend(title='Columns')
plt.tight_layout()
plt.show()
```
